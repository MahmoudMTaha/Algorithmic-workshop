```python
import pandas as pd
import copy
```

```python
def make_knapsack():
    global items
    global max_weight
    items = {}
    item = {}
    item_number = 1
    while True:
        try:
            max_weight =int(input('Please enter the maximum weight of your Knapsack\n').strip())
        except:
            print('Value Error >>> please enter the maximum weight of your Knapsack correctly\n')
        else:
            break
    item_input = input('Please Enter your knapsack items each per line\nin the following format:>> \
    weight  profit\nexample:\n 5 15 \n 1 10\n to finish filling the make_knapsack enter /finish\n')
    while True:
        if item_input == '/finish':
            break
        try:
            item['weight'], item['profit'] = map(int, item_input.split())
            items[str(item_number)] = item.copy()
        except:
            print(f'Value Error >>> Expect two space separated numbers but gets >> {item_input}')
        else:
            item_number += 1
        item_input = input()
    print(f'\nYour Knapsack:\n\n{pd.DataFrame(items)}\n\n with maximum weight: {max_weight}')
```

```python
def knapsack_type_func():
    
    global knapsack_type
    knapsack_types = ['f','1']
    knapsack_type_flag = False
    knapsack_type = input('please enter your knapsacke type:\n for Fractional enter f\n for 1/0 enter 1 \n').strip().lower()
    while not knapsack_type_flag:
        if knapsack_type not in knapsack_types:
            knapsack_type = input('Value Error >>> please enter your knapsacke type correctly\n').strip().lower()
        else:
            knapsack_type_flag = True
```

```python
def cost_fun_base():
    global cost_fun
    cost_fun_types = ['mp','mw','mpw','all']
    flag = False
    cost_fun = input('please enter your cost function type:\n for Maximum Profit enter >> mp\n for Minimum Weight enter >> mw\n for Maximum Profit/Weight enter >> mpw\n for all enter >> all\n').strip().lower()
    while not flag:
        if cost_fun not in cost_fun_types:
            cost_fun = input('Value Error >>> please enter your  cost function type correctly\n').strip().lower()
        else:
            flag = True
```

```python
def view_options():
    print(f'your current selected options are:\n \
    the knapsack items are:\n{pd.DataFrame(items)}\n with maximum size of {max_weight}\n \
    the type of knapsack is : {knapsack_type}\n \
    the cost function type is : {cost_fun}')
```

```python
def menu():
    global status
    menu = ['/fill','/update','/type','/cost','/exit','/options']
    choose = input('please choose what do you want to do:\n \
    to fill the knapsack with current options please enter >> /fill \n \
    to view the current options please enter >> /options \n \
    to Update the objects and the maximum weight of the knapsack please enter >> /update\n \
    to change The type of Knapsack please enter >> /type\n \
    to change The cost function type please enter >> /cost\n \
    to close the program please enter >> /exit\n').strip().lower()
    flag = False
    while not flag:
        if choose not in menu:
            choose = input('Value Error >>> please enter Valid command \n').strip().lower()
        else:
            flag = True
    if choose == '/fill':
        fill_the_knapsack()
    elif choose == '/update':
        make_knapsack()
    elif choose =='/type':
        knapsack_type_func()
    elif choose == '/cost':
        cost_fun_base()
    elif choose == '/options':
        view_options()
    else:
        status = False
        print('GoodBye')
```

```python
def fill_the_knapsack_with_mpw():    
    items_copy = copy.deepcopy(items)
    knapsack_size = max_weight
    knapsack = {}
    for key,value in items_copy.items():
        items_copy[key]['per_unit'] = items_copy[key]['profit'] / items_copy[key]['weight']
    while knapsack_size != 0 : 
        if items_copy == {}:
            break
        max_value_per_unit = max(items_copy, key = lambda x:items_copy[x]['per_unit'])
        weight_of_element = items_copy[max_value_per_unit]['weight']
        if weight_of_element <= knapsack_size:
            knapsack[max_value_per_unit]= copy.deepcopy(items_copy[max_value_per_unit])
            del items_copy[max_value_per_unit]
            knapsack_size -= weight_of_element
        elif knapsack_type == 'f':
            knapsack[max_value_per_unit] = {'weight': knapsack_size}
            knapsack[max_value_per_unit]['profit'] = knapsack_size * items_copy[max_value_per_unit]['per_unit']
            knapsack[max_value_per_unit]['per_unit'] =  items_copy[max_value_per_unit]['per_unit']
            break
        else:
            del items_copy[max_value_per_unit]
    total_Value = 0
    for key in knapsack.keys():
        total_Value += knapsack[key]['profit']
    print(f'the resulting knapsack for Maximum Profit/Weight:\n{pd.DataFrame(knapsack)}\nwith Total Value ={total_Value} \n')    

```

```python
def fill_the_knapsack_with_mp():    
    items_copy = copy.deepcopy(items)
    knapsack_size = max_weight
    knapsack = {}
    while knapsack_size != 0 : 
        if items_copy == {}:
            break
        max_value = max(items_copy, key = lambda x:items_copy[x]['profit'])
        weight_of_element = items_copy[max_value]['weight']
        if weight_of_element <= knapsack_size:
            knapsack[max_value]= copy.deepcopy(items_copy[max_value])
            del items_copy[max_value]
            knapsack_size -= weight_of_element
        elif knapsack_type == 'f':
            knapsack[max_value] = {'weight': knapsack_size}
            knapsack[max_value]['profit'] = knapsack_size * items_copy[max_value]['profit'] / weight_of_element
            break
        else:
            del items_copy[max_value]
            
    total_Value = 0
    for key in knapsack.keys():
        total_Value += knapsack[key]['profit']
    print(f'the resulting knapsack for Maximum Profit:\n{pd.DataFrame(knapsack)}\nwith Total Value ={total_Value} \n')
```

```python
def fill_the_knapsack_with_mw():    
    items_copy = copy.deepcopy(items)
    knapsack_size = max_weight
    knapsack = {}
    while knapsack_size != 0 : 
        if items_copy == {}:
            break
        max_value = min(items_copy, key = lambda x:items_copy[x]['weight'])
        weight_of_element = items_copy[max_value]['weight']
        if weight_of_element <= knapsack_size:
            knapsack[max_value]= copy.deepcopy(items_copy[max_value])
            del items_copy[max_value]
            knapsack_size -= weight_of_element
        elif knapsack_type == 'f':
            knapsack[max_value] = {'weight': knapsack_size}
            knapsack[max_value]['profit'] = knapsack_size * items_copy[max_value]['profit'] / weight_of_element
            break
        else:
            break
    total_Value = 0
    for key in knapsack.keys():
        total_Value += knapsack[key]['profit']
    print(f'the resulting knapsack for Minimum Weight:\n{pd.DataFrame(knapsack)}\nwith Total Value ={total_Value} \n')
```

```python
def fill_the_knapsack():
    if cost_fun == 'mpw':
        fill_the_knapsack_with_mpw()
    elif cost_fun == 'mp':
        fill_the_knapsack_with_mp()
    elif cost_fun == 'mw':
        fill_the_knapsack_with_mw()
    else:
        fill_the_knapsack_with_mpw()
        fill_the_knapsack_with_mp()
        fill_the_knapsack_with_mw()
```

```python
def start():
    print('Welcome\n')
    make_knapsack()
    knapsack_type_func()
    cost_fun_base()
    fill_the_knapsack()
```

```python
def main():
    start()
    while status:
        menu()
```
